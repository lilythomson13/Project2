# Uploading Necessary Files:


```python
from google.colab import files
import pandas as pd

# Upload *merged_tracks_and_albums.csv*
uploaded_merged = files.upload()

# Get the file name (first key in the uploaded dictionary)
file = list(uploaded_merged.keys())[0]

# Load the CSV file
merged_df = pd.read_csv(file, encoding='ISO-8859-1')

# Display first few rows
merged_df.head()
```

```python
print(merged_df.columns)
```

```python

# Define the columns you want to keep
selected_columns = ["album_title", "album_date_released_x","track_genre_top","track_listens"]  # Modify as needed


# Keep only the selected columns
merged_df = merged_df[selected_columns]

```

```python
merged_df.head()
```

```python
merged_df = merged_df.dropna()
merged_df.head()
```

```python
merged_df.head(20)
```

```python
merged_df.rename(columns={"album_date_released_x": "album_release_date"}, inplace=True)
```

```python
print(merged_df.head())
```

```python
merged_df.to_csv("cleaned_merged_tracks.csv", index=False)
```
