# Uploading Necessary Files:


```python
from google.colab import files
import pandas as pd

# Upload *raw_albumns*
uploaded_albums = files.upload()

# Get the file name (first key in the uploaded dictionary)
file1 = list(uploaded_albums.keys())[0]

# Load the CSV file
albums_df = pd.read_csv(file1, encoding='ISO-8859-1')

# Display first few rows
albums_df.head()
```

```python
import pandas as pd

# Upload *tracks*
uploaded_tracks = files.upload()

# Get the file name (first key in the uploaded dictionary)
file3 = list(uploaded_tracks.keys())[0]

# Load the CSV file
tracks_df = pd.read_csv(file3, encoding='ISO-8859-1', header=None)

# Combine first two rows into a single header row
tracks_df.columns = tracks_df.iloc[0] + "_" + tracks_df.iloc[1]  # Merge row 0 and row 1
tracks_df = tracks_df[2:].reset_index(drop=True)  # Remove the first two rows and reset index

# Display cleaned DataFrame
print(tracks_df.head())
```

```python
# Drop the first row after the headers (if it's empty)
tracks_df = tracks_df.iloc[1:].reset_index(drop=True)

# Display the cleaned DataFrame
print(tracks_df.head())
```

```python
# Define the columns you want to keep
selected_columns = ["album_title", "album_date_released", "album_tracks", "artist_location", "artist_name","track_genre_top","track_genres", "track_genres_all","track_listens", "track_title", "album_type"]  # Modify as needed

# Keep only the selected columns
tracks_df = tracks_df[selected_columns]

# Display the result
print(tracks_df.head())
```

cleaning tracks

```python
# Merge the DataFrames on 'album_title'
merged_df = pd.merge(tracks_df, albums_df, on="album_title", how="inner")

# Save the result
merged_df.head(10)

# Save to CSV
merged_df.to_csv("merged_tracks_and_albums.csv", index=False)
```
